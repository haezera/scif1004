# Simulation analysis

```python
import pandas as pd
from InfectiousDiseaseSimulator import InfectiousDiseaseSimulator
import matplotlib.pyplot as plt
```

```python
sim = InfectiousDiseaseSimulator(1000, 1000, 200, 1, 150, 0.01, 0.001, 100)
```

```python
sim.setup()
```

```python
sim.simulate()
```

```python
df = pd.DataFrame(sim.data)
```

```python
plt.plot(df["tick"], df["num_sick"], label="Sick")
plt.plot(df["tick"], df["num_recovered"], label="Recovered")
plt.plot(df["tick"], df["num_healthy"], label="Healthy")
plt.legend()
plt.xlabel("Days since origin")
plt.ylabel("Number of people")
plt.title("Changes in population decomposition")
```
